# Panorama do COVID-19 no Brasil

COVID-19 é uma doença infecciosa causada por recém-descoberto coronavírus.

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

Fato é, não se sabe muita coisa a respeito do COVID-19. Estudos estão sendo realizados no mundo todo, porém os resultados ainda não são conclusivos e definitivos.

<center><img alt="Colaboratory logo" width="50%" src="https://static.poder360.com.br/2022/12/coronavirus-imagem-1-848x477.jpg"></center>

Até o presente momento, observa-se que cerca de 80% dos casos confirmados são assintomáticos e rápidos. A maioria das pessoas que se encaixam nesse grupo, se recupera sem nenhuma sequela.

No entanto, 15% das pessoas terão infecções graves e precisarão de oxigênio. O restante das pessoas, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, irei realizar uma análise sobre os dados públicos da doença.

## Obtenção dos Dados

Os dados utilizados nesta análise exploratoria foram obtidas do repositório público Our World in Data (OWID), que é disponível no GitHub. Este dataset é um conmjunto de informações globais sobre o COVID-19, obtidas por fontes oficiais, governos, organizações de saúde e institutos de pesquisas de pesquisa. Ele inclui variáveis como número de casos confirmados, mortes, taxas de vacinação, dados demográficos e socioeconômicos, como PIB per capita e expectativa de vida, entre outros. Este dataset é atualizado frequentemnete, isso ajuda bastantes na análise dinâmica e atualizada da pandemia.

Com esse conjunto de dados, é possível aprofundar em varias de questões, como:

A evolução temporal da pandemia em diferentes países e regiões.

A relação entre fatores socioeconômicos e o impacto da COVID-19.

A eficácia de políticas públicas e campanhas de vacinação.

A comparação entre países em termos de casos, mortes e medidas de controle.

###Desafios na Análise de uma Doença Nova
Ao analisar uma doença nova igual a covid, abrange uma série de desafios. Por exemplo a falta de dados históricos dificultando a criação de modelos confiáveis e o estudo do comportamento do vírus. Além disso a dinâmica rápida da pandemia exige que os dados sejam atualizados rapidamente, o que pode agravar com varias inconsistências ou demoras nas informações.

Outro desafio significativo são o conflito entre fontes de dados. Os dados oficiais de governos federais e estaduais podem gerar conflito devido a diferenças de métodos utilizados, a demora na notificação de casos e mortes.

Esses conflitos e inconsistências tornam crucial o compartilhamneto de dados, ou seja, a comparação de múltiplas fontes para validar a precisão das informações. A análise crítica dos dados e a contextualização dos resultados são essenciais para evitar conclusões equivocadas.

#O Que Esperar Desse Dataset
Com muitos desafios, o dataset do Our World in Data oferece uma grande quantidade de informaçoes para explorar a pandemia de COVID-19 de forma global. Ele permite identificar tendências, como o impacto das variantes do vírus, a eficácia das vacinas e a relação entre políticas públicas e a evolução da doença.É importante lembrar que os dados refletem apenas o que foi reportado e que erros podem aparecer. Por isso, esta análise foi feita com cautela, sempre considerando o contexto por trás de tudo.

Em resumo, esse dataset é uma ferramenta valiosa para entender a COVID-19, mas sua interpretação exige uma abordagem crítica e contextualizada, levando em conta os desafios da análise de uma doença nova e em constante evolução.


# Dicionário de Variáveis
iso_code	---- Código ISO do país (padrão de 3 letras).

continent	---- Continente onde o país está localizado.

location	---- Nome do país ou região.

date	---- Data da observação (no formato YYYY-MM-DD).

total_cases	---- Número acumulado de casos confirmados de COVID-19 até a data.

new_cases ---- Número de novos casos confirmados reportados na data.

total_deaths ----	Número acumulado de mortes confirmadas por COVID-19 até a
data.

new_deaths ---- Número de novas mortes confirmadas reportadas na data.

total_cases_per_million	---- Número acumulado de casos confirmados por milhão de habitantes.

new_cases_per_million	---- Número de novos casos confirmados por milhão de habitantes na data.

total_deaths_per_million	---- Número acumulado de mortes confirmadas por milhão de habitantes.

new_deaths_per_million	---- Número de novas mortes confirmadas por milhão de habitantes na data.

reproduction_rate	---- Taxa de transmissão (Rt) do vírus na data (média de novas infecções causadas por cada caso).

icu_patients	---- Número de pacientes em unidades de terapia intensiva (UTI) na data.

hosp_patients	---- Número de pacientes hospitalizados devido à COVID-19 na data.

total_tests	Número ---- acumulado de testes realizados até a data.

new_tests	---- Número de novos testes realizados na data.

positive_rate	---- Porcentagem de testes positivos em relação ao total de
testes realizados.

tests_per_case	---- Número de testes realizados por caso confirmado.

total_vaccinations	---- Número acumulado de doses de vacina administradas até a data.

people_vaccinated	---- Número acumulado de pessoas que receberam pelo menos uma dose da vacina.

people_fully_vaccinated	---- Número acumulado de pessoas totalmente vacinadas (com todas as doses necessárias).

new_vaccinations	---- Número de novas doses de vacina administradas na data.

population	---- População total do país ou região.

population_density	---- Densidade populacional (habitantes por km²).

median_age	---- Mediana da idade da população.

aged_65_older	---- Porcentagem da população com 65 anos ou mais.

aged_70_older	---- Porcentagem da população com 70 anos ou mais.

gdp_per_capita	---- PIB per capita (em dólares internacionais).

extreme_poverty	---- Porcentagem da população vivendo em extrema pobreza.

cardiovasc_death_rate	---- Taxa de mortalidade por doenças cardiovasculares (por 100.000 habitantes).

diabetes_prevalence	---- Porcentagem da população com diabetes.

female_smokers	---- Porcentagem de mulheres fumantes.

male_smokers	---- Porcentagem de homens fumantes.

handwashing_facilities	---- Porcentagem da população com acesso a instalações básicas para lavagem das mãos.

hospital_beds_per_thousand	---- Número de leitos hospitalares por mil habitantes.

life_expectancy	---- Expectativa de vida ao nascer.

```
# Encontra a data mais recente no DataFrame
most_recent_date = df['date'].max()

# Filtra o DataFrame para a data mais recente
df_recente = df[df.date == most_recent_date]

# Países com mais casos na data mais recente
países_casos = df_recente[['location', 'total_cases']].sort_values(by='total_cases', ascending=False).head(10)

# Países com mais mortes na data mais recente
países_mortes = df_recente[['location', 'total_deaths']].sort_values(by='total_deaths', ascending=False).head(10)

# Exibir os resultados
print(f"Data mais recente do dataset: {most_recent_date}")
print("\n10 países com mais casos:")
print(países_casos)

print("\n10 países com mais mortes:")
print(países_mortes)
```

Agora vamos plotar um gráfico de barras para análisar 5 países co mais mortes


```
top_5_países_mortes = países_mortes.head(5)

# Plotar gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(top_5_países_mortes['location'], top_5_países_mortes['total_deaths'], color='blue')
plt.title('5 Países com maior número de mortes')
plt.xlabel('Países')
plt.ylabel('Total de mortes')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Análise Exploratória para o Brasil
Agora que ja vimos muito sobre o mundo, vamos ver o Brasil!
* Criar uma cópia do DataFrame apenas com o Brasil
    * `df_brasil = df.loc[df.location == "Brazil"].copy()`
* Identificar quando foi registrada a primeira morte no Brasil
* Identificar quantos dias demorou para registrarmos a primeira morte, considerando o primeiro caso de COVID-19 registrado.
* Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.

```
#cópia do df apenas com o Brasil
df_brasil = df.loc[df.location == "Brazil"].copy()
df_brasil.head()
```

```
df_brasil.date.max()
```

Vimos que nosso dataset mostra as datas desde 31-12-2019 até 15-06-2020.

Vamos olhar quando foram identificados o primeiro caso e a primeira morte no Brasil.

```
df_primeira_morte = df_brasil[df_brasil['total_deaths'] > 0]

# Encontrar a data da primeira morte
data_primeira_morte = df_primeira_morte['date'].min()

print(f"A primeira morte no Brasil foi registrada em: {data_primeira_morte}")
```

```
# Mostrar quantos dias demorou para registrar a primeira morte, levando em conta o primeiro caso de COVID-19 registrado.
# identificaçãso do primeiro caso
primeiro_caso = df_brasil[df_brasil['total_cases'] > 0]['date'].min()

# data da primeira morte
primeira_morte = df_brasil[df_brasil['total_deaths'] > 0]['date'].min()

# Calcular a diferença em dias
calculo_dias_casos_mortes = (primeira_morte - primeiro_caso).days

# Exibir os resultados
print(f"Primeiro caso de COVID-19 no Brasil: {primeiro_caso}")
print(f"Primeira morte por COVID-19 no Brasil: {primeira_morte}")
print(f"Dias entre o primeiro caso e a primeira morte: {calculo_dias_casos_mortes} dias")
```

Agora vamos plotar um gráfico para mostrar a evolução de casos e mortes no Brasil

```
# Plotar gráficos de linha em escalas linear e logarítmica para os casos totais e mortes totais.
x = list(range(168))
z = df_brasil['total_cases']
z2 = df_brasil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9, 7))
ax1.set_title('Casos no Brasil')
ax1.set_xlabel('Dias')
ax1.set_ylabel('Casos')
ax1.plot(x, z)

ax2.set_title('Casos no Brasil (Escala logarítmica)')
ax2.set_xlabel('Dias')
ax2.set_ylabel('Casos')
ax2.set_yscale('log')
ax2.plot(x, z)

ax3.set_title('Mortes no Brasil')
ax3.set_xlabel('Dias')
ax3.set_ylabel('Mortes')
ax3.plot(x, z2)

ax4.set_title('Mortes no Brasil (Escala logarítmica')
ax4.set_xlabel('Dias')
ax4.set_ylabel('Mortes')
ax4.set_yscale('log')
ax4.plot(x, z2)
plt.tight_layout()
```
