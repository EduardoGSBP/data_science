```python
# importando pacotes
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
```

```python


# configurar as cores que usarei no plot
AZUL = "#003F5c"
CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"
```

```python
# dados ficticios
X = ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun", "Jul", "Ago", "Set", "Out", "Nov", "Dez"]
alunos = [155, 172, 160, 173, 161, 137, 202, 187, 123, 210, 117, 172]
projetos = [180, 170, 194, 169, 159, 187, 192, 140, 169, 171, 133, 124]
```

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(9,6), dpi=72)

# plots das linhas
ax.plot(range(12), alunos, color=CINZA)
ax.plot(range(12), projetos, color=AZUL)

# exibir grafico
plt.show()
```

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(9,6), dpi=72, )

# plots das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(0, 300)

# exibir grafico
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9,6), dpi=72, )

# plots das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(0, 300)

# remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar tocos ticks de 12 meses)
ax.set_xticks(ticks=range(12))
ax.set_xticklabels(labels=X)

# exibir grafico
plt.show()
```

Ajuste 3

```python
# definindo a figura
fig, ax = plt.subplots(figsize=(9,6), dpi=72, )

# plots das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

# plots dos scatter plots
ax.scatter(range(12)[-3:], alunos[-3:], s=90, color=CINZA, clip_on=False)
ax.scatter(range(12)[-3:], projetos[-3:], s=90, color=AZUL, clip_on=False)

# anotações nos pontos de scatter
for i, j in enumerate(alunos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=CINZA_TEXTO,
        fontsize=12,
        fontweight='medium'
    )

for i, j in enumerate(projetos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=AZUL,
        fontsize=12,
        fontweight='medium'
    )

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(0, 300)

# remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar tocos ticks de 12 meses)
ax.set_xticks(ticks=range(12))
ax.set_xticklabels(labels=X)

# exibir grafico
plt.show()
```

```python
print("\n\n\n\n")

# definindo a figura
fig, ax = plt.subplots(figsize=(10,6), dpi=72, )

# plots das linhas
ax.plot(range(12), alunos, color=CINZA, linewidth=4)
ax.plot(range(12), projetos, color=AZUL, linewidth=4)

# plots dos scatter plots
ax.scatter(range(12)[-3:], alunos[-3:], s=90, color=CINZA, clip_on=False)
ax.scatter(range(12)[-3:], projetos[-3:], s=90, color=AZUL, clip_on=False)

# anotações nos pontos de scatter
for i, j in enumerate(alunos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=CINZA_TEXTO,
        fontsize=12,
        fontweight='medium'
    )

for i, j in enumerate(projetos[-3:]):
    ax.annotate(
        j,
        xy=(i+9, j),
        xytext=(5, 5),
        textcoords='offset points',
        color=AZUL,
        fontsize=12,
        fontweight='medium'
    )

# legenda no final de cada linha
ax.annotate(
    "Alunos",
    xy=(len(alunos)-1, alunos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=CINZA,
    fontsize=18,
    fontweight='semibold'
            )

ax.annotate(
    "Projetos",
    xy=(len(projetos)-1, projetos[-1]),
    xytext=(50,0),
    textcoords='offset points',
    color=AZUL,
    fontsize=18,
    fontweight='semibold'
            )

# remover spines da direita e superior
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# mudar range do eixo y
ax.set_ylim(100, 220)

# remover yticks e yticklabels
ax.set_yticklabels(labels=[], visible=False)
ax.set_yticks(ticks=[])

# configurar xticks (colocar tocos ticks de 12 meses)
ax.set_xticks(ticks=range(12))
ax.set_xticklabels(labels=X)

# title
ax.text(-0.5, 250,'Taxa de Engajamento', fontsize=24, color=CINZA_TEXTO, fontweight='bold')
ax.text(-0.5, 230,'KPI de engajamento, que considera a quantidade\nde acessos únicos e projetos enviados para\nrevisão  do Sigmoidal', fontsize=13, color=CINZA)

# rodapé
ax.text(
    -0.5,
    80,
    "Fonte: Dados extraídos a partir da Hotmart e Memberkit (21/06/2021)",
    fontsize=12, color=CINZA
    )

# exibir grafico
plt.show()
```
